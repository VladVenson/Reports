```
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Activation, Dropout, Flatten, Dense
#from tensorflow.python.keras.applications import VGG16
from tensorflow.python.keras.optimizers import Adam
import numpy as np
from tensorflow.python.keras.layers import Conv2D, MaxPooling2D
from PIL import Image
import pandas as pd
 

from google.colab import files
```

```
datagen = ImageDataGenerator(rescale=1./255)
train_generator = datagen.flow_from_directory( 
    'tr', color_mode='rgba',
    target_size=(320, 320),
    batch_size=71,
    class_mode= 'categorical',
    shuffle=False)
```

```
datagen_test = ImageDataGenerator(rescale=1./255)
train_generator_test = datagen_test.flow_from_directory(
    'tr_test', color_mode='rgba',
    target_size=(320, 320),
    batch_size=5,
    class_mode= 'categorical',
    shuffle=False)
```

```
(x,y) = train_generator_test.next()

```

```
input_shape = (320, 320, 4)
```

```
model = Sequential()
model.add(Conv2D(128, (4, 4), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (4, 4)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(300))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('sigmoid'))
```

# Новый раздел

```
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```
model.fit_generator(
    train_generator,
    steps_per_epoch=1,   epochs=30,
   # validation_data=train_generator_test,
   # validation_steps = 1
    )
```

```
a = model.predict(x)
a
```
